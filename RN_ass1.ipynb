### Q11: Create a bar chart of student marks with title, xlabel, ylabel

```python
# YOUR CODE HERE
import matplotlib.pyplot as mp

data={
    "x":["python","ML","DS","AI","DBMS"],
    "y":[60,70,80,90,85]
}

x=data["x"]
y=data["y"]

mp.title("Student performance")
mp.xlabel("Subject")
mp.ylabel("Marks")

mp.bar(x,y)
mp.show()
```

### Q12: Create a pie chart of time spent

```python
import matplotlib.pyplot as mp

data = {
    "x": ["coding", "study", "sports", "sleep"],
    "y": [20, 30, 10, 40]
}

mp.pie(data["y"], labels=data["x"], autopct='%1.1f%%')
mp.title("Daily Activity Distribution")
mp.show()
```
